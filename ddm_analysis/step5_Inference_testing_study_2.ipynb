# Test Main Effect and Interaction Effect

```python
# load the model fitted in step XX
m_FULL_hypothesis = hddm.load("study2_regression_full_hypothesis")
```

```python
# get the MCMC fitting nodes of each parameters in the model
s2_full_v_movie_valence,\
s2_full_v_mood_valence,\
s2_full_v_movie_arousal,\
s2_full_v_mood_arousal,\
s2_full_v_movie_valence_mood_valence,\
s2_full_v_movie_arousal_mood_arousal,\
s2_full_a_movie_valence,\
s2_full_a_mood_valence,\
s2_full_a_movie_arousal,\
s2_full_a_mood_arousal,\
s2_full_a_movie_valence_mood_valence,\
s2_full_a_movie_arousal_mood_arousal,\
s2_full_t_movie_valence,\
s2_full_t_mood_valence,\
s2_full_t_movie_arousal,\
s2_full_t_mood_arousal,\
s2_full_t_movie_valence_mood_valence,\
s2_full_t_movie_arousal_mood_arousal,\
s2_full_z_movie_valence,\
s2_full_z_mood_valence,\
s2_full_z_movie_arousal,\
s2_full_z_mood_arousal,\
s2_full_z_movie_valence_mood_valence,\
s2_full_z_movie_arousal_mood_arousal = m_FULL_hypothesis.nodes_db.loc[[
    "v_movie_valence[T.1]",
    "v_mood_valence[T.1]",
    "v_movie_arousal[T.1]",
    "v_mood_arousal[T.1]", 
    "v_movie_valence[T.1]:mood_valence[T.1]",
    "v_movie_arousal[T.1]:mood_arousal[T.1]",
    
    "a_movie_valence[T.1]",
    "a_mood_valence[T.1]",
    "a_movie_arousal[T.1]",
    "a_mood_arousal[T.1]", 
    "a_movie_valence[T.1]:mood_valence[T.1]",
    "a_movie_arousal[T.1]:mood_arousal[T.1]",
    
    "t_movie_valence[T.1]",
    "t_mood_valence[T.1]",
    "t_movie_arousal[T.1]",
    "t_mood_arousal[T.1]", 
    "t_movie_valence[T.1]:mood_valence[T.1]",
    "t_movie_arousal[T.1]:mood_arousal[T.1]",
    
    "z_movie_valence[T.1]",
    "z_mood_valence[T.1]",
    "z_movie_arousal[T.1]",
    "z_mood_arousal[T.1]", 
    "z_movie_valence[T.1]:mood_valence[T.1]",
    "z_movie_arousal[T.1]:mood_arousal[T.1]"], 'node']
```

```python
# test the percentage of posteriors higher and lower than 0
def inference_test(trace):
    probability1 = round((trace < 0).mean(), 3) 
    probability2 = round((1-probability1), 3) 
    return "{} % < 0 < {} %".format(probability1 * 100, probability2 * 100)
```

## drift rate hypothesis testing

```python
# test the main effect of valence and arousal against 0
print(inference_test(s2_full_v_movie_valence.trace()))
print(inference_test(s2_full_v_movie_arousal.trace()))
print(inference_test(s2_full_v_mood_valence.trace()))
print(inference_test(s2_full_v_mood_arousal.trace()))

# test the interaction between movie and mood factors for valence and arousal
print(inference_test(s2_full_v_movie_valence_mood_valence.trace()))
print(inference_test(s2_full_v_movie_arousal_mood_arousal.trace()))

# compare the size of interaction effect of valence and arousal
print(inference_test(s2_full_v_movie_valence_mood_valence.trace() - s2_full_v_movie_arousal_mood_arousal.trace()))
```

```python
# get the statistics of the posteior,
# including the mean of posterior and the 0.025 & 0.975 quantiles to get the 0.95 credible intervals
s2_full_stats = m_FULL_hypothesis.gen_stats()
s2_full_stats[s2_full_stats.index.isin(["v_Intercept",
    "v_movie_valence[T.1]",
    "v_mood_valence[T.1]",
    "v_movie_arousal[T.1]",
    "v_mood_arousal[T.1]", 
    "v_movie_valence[T.1]:mood_valence[T.1]",
    "v_movie_arousal[T.1]:mood_arousal[T.1]",])]
```

## Decison boundary hypothesis testing

```python
# test the main effect of valence and arousal against 0
print(inference_test(s2_full_a_movie_valence.trace()))
print(inference_test(s2_full_a_movie_arousal.trace()))

print(inference_test(s2_full_a_mood_valence.trace()))
print(inference_test(s2_full_a_mood_arousal.trace()))

# test the interaction between movie and mood factors for valence and arousal
print(inference_test(s2_full_a_movie_valence_mood_valence.trace()))
print(inference_test(s2_full_a_movie_arousal_mood_arousal.trace()))
```
